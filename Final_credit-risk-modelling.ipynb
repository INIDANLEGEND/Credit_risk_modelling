```python
# now we will check how many are categorical columns 

cat_columns=[]
for i in df.columns:
    if df[i].dtype == 'object':
        cat_columns.append(i)
        
# so we have 5 categorical columns
```

**chi2 Test**

chi2 test is basically performed between two categorical columns to check whether one is dependent on other or not. we calculate p value using chi2 test and if it comes greater than significance value then we say we cann't reject the H0 and hence the two columns arre not associated but if it is less than or equal to significance level then we say that we reject H0 and hence two columns are associated.


significance level is generally taken as 5 percent or 0.05 which means we have 95 percent confidence and 5 percent we can take risk


H0(NULL Hypothesis) : not associated


H1(Alternate Hypothesis) : associated


```python
from scipy.stats import chi2_contingency

# so we will perform chi-square test between categorical columns
for i in cat_columns:
    chi2,pval,_,_ = chi2_contingency(pd.crosstab(df[i],df['Approved_Flag']))
    print(i,'---',pval)
    
# we can see that all the values we are getting is less than 0.05 percent and hence we can say that we reject H0 which means two columns are associated
```

```python
numeric_columns=[]
for i in df.columns:
    if df[i].dtype!='object' and i not in ['PROSPECTID','Approved_Flag']:
        numeric_columns.append(i)
```

**Multicollinearity**

here we will check for multicollinearity which means whether we can predict that column using a linear regression model on other columns if yes we should remove that column as it will not add any extra inputor vaue to our model . we calculate multicollinearity using VIF and if it gives a high value generally greater than 5 or 10 we say taht we can predict that column using the other columns and hence we should remove that column.

There are Two methods for VIF : Parallel and Sequential


Parallel : calculate VIF for all columns without removing any column and after calculating all VIF values drop column with high VIF value. This method is not good becaues suppose c1 and c2 are related so we remove both of them but we should remove only one of them

Sequential : calculate vif for one column and then remove it if it is higher than threshold hence we will keep the one of the c1 or c2 and not remove both as was the above case.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = df[numeric_columns]
total_columns=vif_data.shape[1]
columns_to_be_kept = []
column_index=0

# Calculate vif value
for i in range(0,total_columns):
    vif_value = variance_inflation_factor(vif_data,column_index)
    print(numeric_columns[i],'---',vif_value)
    
    if vif_value<=6:
        columns_to_be_kept.append(numeric_columns[i])
        column_index=column_index+1
    else:
        vif_data = vif_data.drop([numeric_columns[i]],axis=1)
```

**Anora Test**

this test is used to calculate the relationship between numerical and categorical columns (Greater than 3 categories). here also similar to the chi2 test we say that two columns are related if p-value using the anora test is less than or equal to 0.05 else we say that the two columns are not related.

Here we are using one way anora test

```python
# now we will check which numerical columns are related when predicting approved flag
# for numerical column vs categorical(more than 2 categories) we use Anora Test

from scipy.stats import f_oneway
columns_to_be_kept_numerical = []
for i in columns_to_be_kept:
    a=list(df[i])
    b=list(df['Approved_Flag'])
    
    group_P1 = [value for value, group in zip(a, b) if group == 'P1']
    group_P2 = [value for value, group in zip(a, b) if group == 'P2']
    group_P3 = [value for value, group in zip(a, b) if group == 'P3']
    group_P4 = [value for value, group in zip(a, b) if group == 'P4']
    
    f_statistics,p_value = f_oneway(group_P1,group_P2,group_P3,group_P4)
    
    if p_value<=0.05:
        columns_to_be_kept_numerical.append(i)
```

```python
features = columns_to_be_kept_numerical + ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']
df = df[features + ['Approved_Flag']]
df

# now df dataset contains our final dataframe which we will further use for model building
```

```python
# now we will encode the categorical columns 

df['MARITALSTATUS'].unique()    
df['EDUCATION'].unique()
df['GENDER'].unique()
df['last_prod_enq2'].unique()
df['first_prod_enq2'].unique()

# we will perform ordinal encoding for Education and one-hot encoding for other categories
df.loc[df['EDUCATION']=='SSC',['EDUCATION']]=1
df.loc[df['EDUCATION'] == '12TH',['EDUCATION']]             = 2
df.loc[df['EDUCATION'] == 'GRADUATE',['EDUCATION']]         = 3
df.loc[df['EDUCATION'] == 'UNDER GRADUATE',['EDUCATION']]   = 3
df.loc[df['EDUCATION'] == 'POST-GRADUATE',['EDUCATION']]    = 4
df.loc[df['EDUCATION'] == 'OTHERS',['EDUCATION']]           = 1
df.loc[df['EDUCATION'] == 'PROFESSIONAL',['EDUCATION']]     = 3

df['EDUCATION'].value_counts()
df['EDUCATION'] = df.loc[:,'EDUCATION'].astype(int)
df.info()
```

```python
df_encoded = pd.get_dummies(df, columns=['MARITALSTATUS','GENDER', 'last_prod_enq2' ,'first_prod_enq2'])
df_encoded.info()
df_encoded.describe()
```

**Random Forest**

```python
# now we have pre-processed the dataframe and will move to model training part

# first we will try with Random Forest

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, precision_recall_fscore_support
from sklearn.preprocessing import LabelEncoder

y = df_encoded['Approved_Flag']
x = df_encoded.drop(['Approved_Flag'],axis=1)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

x_train,x_test,y_train,y_test = train_test_split(x,y_encoded,test_size=0.2,random_state=42)

# random state parameter har baar same result or split produce ho isiliye karte hai

rf_classifier= RandomForestClassifier(n_estimators = 200,random_state=42)
rf_classifier.fit(x_train,y_train)
y_pred=rf_classifier.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy : {accuracy}')
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred)

for i,v in enumerate(['P1',"P2",'P3','P4']):
    print(f"class : {v}")
    print(f"precision : {precision[i]}")
    print(f"recall : {recall[i]}")
    print(f"F1 Score : {f1_score[i]}")
```

```python
# hence we can see that we are getting better accuracy, precision and recall values when using xgboost 
# now before changing its parameters we will first try to scale all columns to same level so that 
# one column will not have much effect on the model

from sklearn.preprocessing import StandardScaler

for i in x.columns:
    column_data = df_encoded[i].values.reshape(-1,1)
    scaler = StandardScaler()
    scaled_column = scaler.fit_transform(column_data)
    df_encoded[i] = scaled_column
    
df_encoded
```

```python
# now we will again apply xgboost after standardising all the columns

import xgboost as xgb
from sklearn.preprocessing import LabelEncoder


xgb_classifier = xgb.XGBClassifier(objective='multi:softmax',num_class=4)

y=df_encoded['Approved_Flag']
x=df_encoded.drop(['Approved_Flag'],axis=1)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

x_train,x_test,y_train,y_test=train_test_split(x,y_encoded,test_size=0.2,random_state=42)

xgb_classifier.fit(x_train,y_train)
y_pred = xgb_classifier.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy : {accuracy}')
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred)

for i,v in enumerate(['P1',"P2",'P3','P4']):
    print(f"class : {v}")
    print(f"precision : {precision[i]}")
    print(f"recall : {recall[i]}")
    print(f"F1 Score : {f1_score[i]}")
```

```python
# we got the same result even after applying standardization 
# so we will try to tune it with some hyper parameters and see whether we can get better result than this or not

# so we will be using the grid Search Cv library for finding the best hyperparameters among some given hyperparameters

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
}

xgb_model = xgb.XGBClassifier(objective='multi:softmax',num_class=4)

grid_search = GridSearchCV(estimator=xgb_model,param_grid=param_grid,cv=3,scoring='accuracy',n_jobs=-1)
grid_search.fit(x_train,y_train)

print("Best HyperParameters : ",grid_search.best_params_)

best_model = grid_search.best_estimator_
accuracy = best_model.score(x_test,y_test)
print("Test Accuracy : ",accuracy)
```
